# Assignment 1
1. Create a JSON file (employee.json) containing employee information of minimum 5 employees.
Each employee information consists of Name, DOB, Height, City, State. 
Write a python program that reads this information from the JSON file and saves the information into a list of objects of 
Employee class. Finally print the list of the Employee objects.

2. Create a dictionary of any 7 Indian states and their capitals. Write this into a JSON file.

```python
# TASK 1:
import json
class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state
employee_data = '''
{
  "employees": [
    {
      "name": "Rahul Goel",
      "dob": "1990-05-15",
      "height": 175,
      "city": "Mumbai",
      "state": "Maharashtra"
    },
    {
      "name": "Jane Smith",
      "dob": "1985-09-28",
      "height": 162,
      "city": "Los Angeles",
      "state": "California"
    },
    {
      "name": "Sita Sharma",
      "dob": "1992-03-12",
      "height": 180,
      "city": "Madurai",
      "state": "Tamil Nadu"
    },
    {
      "name": "Komal Naik",
      "dob": "1994-11-03",
      "height": 168,
      "city": "Akola",
      "state": "Maharashtra"
    },
    {
      "name": "Hari Parmar",
      "dob": "1999-07-19",
      "height": 175,
      "city": "Ahemdabad",
      "state": "Gujrat"
    }
  ]
}
'''

data = json.loads(employee_data)

employees = []
for emp in data["employees"]:
    employee = Employee(emp["name"], emp["dob"], emp["height"], emp["city"], emp["state"])
    employees.append(employee)

for employee in employees:
    print("Name:", employee.name)
    print("DOB:", employee.dob)
    print("Height:", employee.height)
    print("City:", employee.city)
    print("State:", employee.state)
    print()

output_data = {"employees": []}
for employee in employees:
    output_data["employees"].append({
        "name": employee.name,
        "dob": employee.dob,
        "height": employee.height,
        "city": employee.city,
        "state": employee.state
    })

with open("employee.json", "w") as file:
    json.dump(output_data, file)
```

# Assignment 2
 1. Create a class named ‘Dog’. It should have a constructor which accepts its name, age and coat color. You must perform the following operations:
a. It should have a function ‘description()’ which prints the name and age of the dog.
b. It should have a function ‘get_info()’ which prints the coat color of the dog.
c. Create child classes ‘JackRussellTerrier’ and ‘Bulldog’ which is inherited from the class ‘Dog’. It should have at least two methods of its own.
d. Create objects and implement the above functionalities.

```python
class Dog:
    def __init__(self, name, age, coat_color):
        self.name = name
        self.age = age
        self.coat_color = coat_color
    
    def description(self):
        print("Name:", self.name)
        print("Age:", self.age)
    
    def get_info(self):
        print("Coat Color:", self.coat_color)

class JackRussellTerrier(Dog):
    def __init__(self, name, age, coat_color):
        super().__init__(name, age, coat_color)
    
    def bark(self):
        print("Jack Russell Terrier is barking!")

class Bulldog(Dog):
    def __init__(self, name, age, coat_color):
        super().__init__(name, age, coat_color)
    
    def run(self):
        print("Bulldog is running!")

dog1 = Dog("Buddy", 3, "Brown")
dog1.description()
dog1.get_info()
print()

dog2 = JackRussellTerrier("Max", 5, "White and Brown")
dog2.description()
dog2.get_info()
dog2.bark()
print()

dog3 = Bulldog("Rocky", 4, "Brown")
dog3.description()
dog3.get_info()
dog3.run()
```
